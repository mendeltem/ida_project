```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np

np.random.seed(42)
#generating number between 0 and 10
rvs =binom.rvs(n=10, p=0.5, size = 100)
```

```python
plt.plot(rvs)
```

```python
from scipy.stats import bernoulli
import numpy as np

np.random.seed(42)
#generating number 0 and 1
flips=  bernoulli.rvs(p=0.5, size = 100)
```

```python
plt.plot(flips)
```

```python
#prob of 5 heads after 10 throes with a fair coin
binom.pmf(k=5, n=10, p=0.5)
```

```python
#prob of 5 heads or less after 10 throws with p = 0.5 
binom.cdf(k=5,n=10,p=0.5) 
```

```python
#prob of 5 heads or higher after 10 throws with p = 0.5 
1 - binom.cdf(k=5,n=10,p=0.5) 
```
