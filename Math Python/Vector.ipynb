```python
from fractions import Fraction
import matplotlib.pyplot as plt
import numpy as np



from random import randint
import math
```

```python
def plot_vector(start, vector):
    plt.figure()
    ax = plt.gca()
    ax.set_facecolor('xkcd:salmon')
    ax.set_facecolor((0.0, 0.97, 0.42))
    ax.quiver(start[0], start[1], vector[0], vector[1], 
              
              color = "r",
              angles='xy', 
              scale_units='xy', 
              scale=1,animated=True)
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])
    plt.draw()
    plt.show()
       
start = np.array([0,0])
vector = np.array([2,2])
print("start point",start)
print("vector point",vector)
plot_vector(start, vector)    
```

```python
vector_1 = vector * -2
plot_vector(start, vector_1)
```

```python
matrix = np.array([[-1,0],[0,1]])
print("Matrix: \n", matrix)
vector_1 = np.dot(matrix,vector)
plot_vector(start, vector_1)
```

```python
matrix = np.array([[1,0],[0,1]])
print("Matrix: \n", matrix)
vector_1 = np.dot(matrix,vector)
plot_vector(start, vector_1)
```
